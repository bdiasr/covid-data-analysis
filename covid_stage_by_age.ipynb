# Gráfico da taxa de mortos pela idade

```python
plt.bar(ageList, studyListDeath, label="Óbitos", color='red')
plt.title("Taxa de óbitos pela COVID-19 por idade")
plt.xlabel("Idade")
plt.ylabel("Número de mortes")

plt.legend()
plt.show()
```

# Relação de internados e internados na UTI por idade

```python
plt.bar(ageList, studyList_Hospitalized, label="Internados")
plt.bar(ageList, studyList_HospitalizedICU, label="Internados na UTI")
plt.title("Taxa de Internação da COVID-19 por idade")
plt.xlabel("Idade")
plt.ylabel("Número de mortes")
plt.legend()
plt.show()
```

#Relação de cura por idade

```python
plt.bar(ageList, studyListCure, label="Quantidade de pacientes curados", color='green')
plt.title("Taxa de Cura da COVID-19 por idade")
plt.xlabel("Idade")
plt.ylabel("Número de pacientes curados por idade")
plt.legend()
plt.show()
```

# Relação de mortes, internação e óbito por idade

```python
fig, ax = plt.subplots(1, 2, sharex=False, figsize=(10,10))
ax[0].bar(ageList, studyListCure, label = "Pacientes Curados", color='green')
ax[0].set_title('Relação de curados')
ax[0].set_xlabel('Idade')
ax[0].set_ylabel("Estado do Paciente")

ax[1].bar(ageList, studyList_Hospitalized, label = "Internado")
ax[1].bar(ageList, studyList_HospitalizedICU, label="Internado na UTI")
ax[1].bar(ageList, studyListDeath, label="Óbito", color='red')
ax[1].set_title('Relação de mortes, internação e óbito por idade')
ax[1].set_xlabel('Idade')
ax[1].set_ylabel("Estado do Paciente")

fig.legend(loc='upper left')
fig.autofmt_xdate(rotation=90)

plt.show()
```
